```python
# Remove ₹, commas, and extra spaces
df["Monthly_Bill"] = df["Monthly_Bill"].str.replace("₹", "").str.replace(",", "").str.strip()
```

```python
import numpy as np

def parse_price_range(x):
    x = x.strip()
    if "–" in x:  # range like 500 – 1000
        parts = x.split("–")
        low = int(parts[0].strip())
        high = int(parts[1].strip())
        mid = (low + high) / 2
        return low, high, mid
    
    elif x.lower().startswith("below"):  # e.g. Below 500
        high = int(x.split()[-1])
        return 0, high, high/2   # assume min=0, mid=high/2
    
    elif x.lower().startswith("above"):  # e.g. Above 5000
        low = int(x.split()[-1])
        return low, 10000, low   # no upper bound, set max=NaN, mid=low (or leave NaN)
    
    else:
        return np.nan, np.nan, np.nan

df[["min_price","max_price","mid_price"]] = df["Monthly_Bill"].apply(lambda x: pd.Series(parse_price_range(x)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Distribution of residence types
plt.figure()
sns.countplot(data=data, x="Type_of_Residence", palette="Set2")
plt.title("Distribution of Residence Types")
plt.ylabel("Count")
plt.xticks(rotation=30)
plt.show()

# Most responses are from Hostel/PG and Apartments.
```

```python
 # Household size distribution by residence type

plt.figure()
sns.boxplot(data=data, x="Type_of_Residence", y="No_of_People_in_Household", palette="Set3")
plt.title("Household Size by Residence Type")
plt.xticks(rotation=30)
plt.show()
# PG/Hostel tends to have 2–3 members, while independent houses are larger.
```

```python
 # Average electricity cost by area

avg_area_cost = data.groupby("Area_in_chennai")["mid_price"].mean().sort_values(ascending=False)
plt.figure()
sns.barplot(x=avg_area_cost.values, y=avg_area_cost.index, palette="coolwarm")
plt.title("Areas by Average Electricity Cost")
plt.xlabel("Average Price (₹)")
plt.show()

# Some areas have significantly higher bills (₹1500–₹2000).
```

```python
# Power cut frequency vs duration

plt.figure(figsize=(10,5))
sns.countplot(data=data, x="Power_Cut_Frequency", hue="Duration_of_power_cut", palette="Set1")
plt.title("Power Cut Frequency vs Duration")
plt.ylabel("Duration")
plt.xticks(rotation=30)
plt.show()
# In “Weekly” cuts, many face up to 1–2 hours of outage.
```

```python
# Impact of power cuts on households

plt.figure()

impact_counts = data["PowerCut_Impact"].value_counts()

plt.pie(impact_counts, labels=impact_counts.index, autopct="%1.1f%%",
        startangle=90, colors=sns.color_palette("pastel"), wedgeprops=dict(width=0.4))

plt.title("Impact of Power Cuts on Households")
plt.show()
# Majority report major impact.
```

```python
# TNEB service rating distribution

plt.figure()
sns.histplot(data["TNEB_service_ratings"], bins=5, kde=True, color="skyblue")
plt.title("Distribution of TNEB Service Ratings")
plt.xlabel("Rating (1-5)")
plt.ylabel("Rating Level")
plt.show()
# 3–4 ratings, showing mixed satisfaction.
```

```python
# Alternative power solutions adoption

plt.figure()
data["Alt_Power_Solution"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Alternative Power Solutions")
plt.ylabel("")
plt.show()
# Many households show interest in backup (Yes/Maybe > No).
```

```python
# AC/Heater usage vs electricity cost

plt.figure()
sns.boxplot(data=data, x="Hrs_of_AC/Heater_usage_per_day", y="mid_price", palette="muted")
plt.title("AC/Heater Usage vs Electricity Cost")
plt.xlabel("Hours of AC/Heater usage per day")
plt.ylabel("Bill Price (₹)")
plt.show()
# more AC usage → higher bills.
```
